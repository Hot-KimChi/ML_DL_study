```python
import os

fname = os.path.join("./jena_data/jena_climate_2009_2016.csv")

with open(fname) as f:
    data = f.read()

lines = data.split("\n")
header = lines[0].split(",")

lines = lines[1:]

print(header)
print(len(header))
print(len(lines))
```

>#### 데이터 전체를 numpy 배열로 변경.

```python
import numpy as np

temperature = np.zeros(len(lines),)
print("온도 데이터 array:", temperature.shape)

raw_data = np.zeros((len(lines), len(header) - 1)) ## Date Time 열 제외하기 위해(14개 데이터만 사용)
print("미래 온도를 예측하기 위해 사용할 feature:",raw_data.shape)

for i, line in enumerate(lines):
    values = list(float(x) for x in line.split(",")[1:]) ## 첫번째 열[Date Time]을 빼고 전달.
    temperature[i] = values[1]   ## temperature 배열 저장
    raw_data[i, :] = values[:]   ## 온도 포함하여 모든 열을 저장 
```

```python
from matplotlib import pyplot as plt

plt.plot(range(len(temperature)), temperature)
plt.show()
```

>#### 처음 10일 간 온도데이터 그리기(10분마다 데이터 sampling)

```python
plt.plot(range(10*144), temperature[:1440])
plt.show()
```

- 하루하루 데이터를 살펴보면, 온도 변화는 매우 불안정

>#### 검증세트와 테스트세트 분할하여 예측하기

- Train set: 50%
- validation set: 25%
- test set: 25%

```python
train_cnt = int(len(temperature)*0.5)
print(train_cnt)

val_cnt = int(len(temperature)*0.25)
print(val_cnt)

test_cnt = len(temperature) - train_cnt - val_cnt
print(test_cnt)

print("Total count:", train_cnt + val_cnt + test_cnt)
```

>### 10.2.1_데이터 준비
> Target: 한시간에 한번씩 샘플링된 5일간의 데이터가 주어졌을 때, 24시간 뒤의 온도를 예측할 수 있나

```python
## 데이터 정규화
mean = raw_data[:train_cnt].mean(axis=0)
raw_data = raw_data - mean

std = raw_data[:train_cnt].std(axis=0)
raw_data = raw_data / std
```

>#### 훈련, 검증, 테스트 데이터셋 만들기
- 시간 당 하나의 데이터 포인트의 샘플링으로 변환(sampling_rate = 6)
- 5일간(120시간) 데이터를 사용(sequence_lenth = 120)

```python
sampling_rate = 6
sequence_length = 120
delay = sampling_rate * (sequence_length + 24 - 1) ## 시퀸스 끝에서 24시간 후의 온도 

batch_size = 256

import keras

train_dataset = keras.utils.timeseries_dataset_from_array(
    raw_data[:-delay],
    targets = temperature[delay:],
    sampling_rate = sampling_rate,
    sequence_length = sequence_length,
    shuffle = True,
    batch_size = batch_size,
    start_index = 0,
    end_index = train_cnt)

val_dataset = keras.utils.timeseries_dataset_from_array(
    raw_data[:-delay],
    targets = temperature[delay:],
    sampling_rate = sampling_rate,
    sequence_length = sequence_length,
    shuffle = True,
    batch_size = batch_size,
    start_index = train_cnt,
    end_index = train_cnt + val_cnt)

test_dataset = keras.utils.timeseries_dataset_from_array(
    raw_data[:-delay],
    targets = temperature[delay:],
    sampling_rate = sampling_rate,
    sequence_length = sequence_length,
    shuffle = True,
    batch_size = batch_size,
    start_index = train_cnt + val_cnt)    
```

```python
## 훈련 데이터셋의 배치 크기 확인하기
for samples, targets in train_dataset.take(1):  ## 1번째 batch size 확인하기.
    print("샘플 크기:", samples.shape)
    print("target size:", targets.shape)   
```

>### 10.2.2_상식 수준의 기준점
- 딥러닝 모델을 사용하여 온도 예측 문제를 풀기 전에 간단한 상식 수준의 해법 시도
- 이유는 정상적인 문제인지 확인하기 위한 용도이며, 고수준 머신 러닝 모델이라면 뛰어 넘어야 할 기준점이 된다.

```python
## 상식 수준 모델의 MAE 계산하기
def evaluate_naive_method(dataset):
    total_abs_err = 0.
    samples_seen = 0
    for samples, targets in dataset:
        prediction = samples[:, -1, 1] * std[1] + mean[1]
        total_abs_err += np.sum(np.abs(prediction - targets))
        samples_seen += samples.shape[0]
    return total_abs_err / samples_seen

print(f"val MAE: {evaluate_naive_method(val_dataset): .2f}")
print(f"test MAE: {evaluate_naive_method(test_dataset): .2f}")
```

>### 10.2.3_기본적인 머신 러닝 모델 시도해 보기
> 상식 수준의 기준점 설계한 이 후에 손쉽게 만들 수 있는 머신 러닝 모델 설계
- 2개 Dense층을 통과하는 완전 연결 네트워크 설계
- 회귀문제이기에 마지막 Dense층에 활설화 함수를 두지 않음
- 손실함수로 MAE대신 MSE(평균제곱오차) 사용

```python
print(raw_data.shape[-1])
```

```python
from tensorflow import keras
from tensorflow.keras import layers

inputs = keras.Input(shape=(sequence_length, raw_data.shape[-1]))
x = layers.Flatten()(inputs)
x = layers.Dense(16, activation="relu")(x)
outputs = layers.Dense(1)(x)

model = keras.Model(inputs, outputs)

callbacks = [
    keras.callbacks.ModelCheckpoint("jena_dense.keras",
                                    save_best_only=True)
]

model.compile(optimizer="rmsprop", loss="mse", metrics=["mae"])
history_full = model.fit(train_dataset,
                    epochs=10,
                    validation_data=val_dataset,
                    callbacks=callbacks)                                                        
```

```python
## 최상의 모델의 로드하여 테스트데이트로 평가
model = keras.models.load_model("jena_dense.keras")
print(f"test MAE: {model.evaluate(test_dataset)[1]:.2f}")
```

```python
## 결과 그래프 그리기
import matplotlib.pyplot as plt

def plot(history):
    loss = history.history["mae"]
    val_loss = history.history["val_mae"]
    epochs = range(1, len(loss) + 1)
    
    plt.figure()
    plt.plot(epochs, loss, "bo", label="Training MAE")
    plt.plot(epochs, val_loss, "b", label="Validation MAE")
    plt.title("Training and validation MAE")
    plt.legend()
    plt.show()

plot(history_full)
```

>### 10.2.4_1D 합성곱 모델 시도해보기
- 좋은 feature engineering과 더불어 아키텍처 구조를 활용하자.
- 한번에 24시간의 데이터를 sampling하려면, Maxpooling1D층 윈도우 크기 = 24

```python
inputs = keras.Input(shape=(sequence_length, raw_data.shape[-1]))
x = layers.Conv1D(8, 24, activation="relu")(inputs)
x = layers.MaxPooling1D(2)(x)
x = layers.Conv1D(8, 12, activation="relu")(x)
x = layers.MaxPooling1D(2)(x)
x = layers.Conv1D(8, 6, activation="relu")(x)
x = layers.GlobalAveragePooling1D()(x)
outputs = layers.Dense(1)(x)

model = keras.Model(inputs, outputs)

callbacks = [
    keras.callbacks.ModelCheckpoint("jena_conv.keras",
                                    save_best_only=True)
]

model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])

history = model.fit(train_dataset,
                    epochs=10,
                    validation_data=val_dataset,
                    callbacks=callbacks)
```

```python
model = keras.models.load_model('jena_conv.keras')
print(f"test MAE: {model.evaluate(test_dataset)[1]:.2f}")
```

>### 10.2.5_첫번째 순환 신경망
- 인과관게와 순서가 의미있는 시퀀스 데이터를 사용하여 LSTM 층을 사용

```python
inputs = keras.Input(shape=(sequence_length, raw_data.shape[-1]))
x = layers.LSTM(16)(inputs)
outputs = layers.Dense(1)(x)

model = keras.Model(inputs, outputs)

callbacks = [
    keras.callbacks.ModelCheckpoint('jena_lstm.keras',
                                    save_best_only=True)
]

model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])

history_lstm = model.fit(train_dataset,
                         epochs=10,
                         validation_data=val_dataset,
                         callbacks=callbacks)
```

```python
plot(history_lstm)
```

```python
model = keras.models.load_model('jena_lstm.keras')
print(f'테스트 MAE: {model.evaluate(test_dataset)[1]:.2f}')
```

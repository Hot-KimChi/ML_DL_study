>### 1) 잔차연결
- 함수 연결이 너무 깊으면 잡음이 정보를 추월
- 역전파가 동작하지 않게 동작(모델이 훈련이 되지 않음)
- 그레이디언트 소실 발생
- How
    - 층이나 블록의 입력을 출력에 더하기만 하면 됨
    - relu 활성화 함수나 드롭아웃 층을 돌아 가게 설계
    - 입력을 블록의 출력에 다시 더하기 때문에 출력 크기가 입력과 같아야 한다.
- 잔차 연결을 사용하면, 그레이디언트 소실 걱정없이 원하는 깊이의 네트워크 설계 가능

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
## 필터 개수가 변경되는 잔차 블록
def get_model1():

    inputs = keras.Input(shape=(32, 32, 3))

    ## 잔차를 따로 저장
    x = layers.Conv2D(32, 3, activation='relu', padding='same')(inputs)
    residual = x

    ## 출력 필터 32 --> 64 증가 / same padding으로 다운샘플링 방지
    x = layers.Conv2D(64, 3, activation='relu', padding='same')(x)

    ## 잔차가 32개 필터이기에 동일하게 하기 위해 잔차 층 수정.
    residual = layers.Conv2D(64, 1)(residual)

    x = layers.add([x, residual])

    outputs = layers.Dense(1, activation='sigmoid')(x)

    model = keras.Model(inputs, outputs)

    return model


model = get_model1()
model.summary()
keras.utils.plot_model(model, show_shapes=True)
```

```python
## 최대 풀링 층을 가진 잔차 블록
def get_model2():

    inputs = keras.Input(shape=(32, 32, 3))

    ## 잔차를 따로 저장
    x = layers.Conv2D(32, 3, activation='relu', padding='same')(inputs)
    residual = x

    ## 출력 필터 32 --> 64 증가 / same padding으로 다운샘플링 방지
    x = layers.Conv2D(64, 3, activation='relu', padding='same')(x)
    x = layers.MaxPooling2D(2, padding='same')(x)

    ## 잔차가 32개 필터이기에 동일하게 하기 위해 잔차 층 수정.
    ## MaxPooling2D 으로 크기 조정을 strides로 같게 수정.
    residual = layers.Conv2D(64, 1, strides=2)(residual)

    x = layers.add([x, residual])

    outputs = layers.Dense(1, activation='sigmoid')(x)

    model = keras.Model(inputs, outputs)

    return model


model = get_model2()
model.summary()
keras.utils.plot_model(model, show_shapes=True)
```

>### 간단한 컨브넷 구성

```python
inputs = keras.Input(shape=(32, 32, 3))

## nomalization
x = layers.Rescaling(1./255)(inputs)

def residual_block(x, filters, pooling=False):
    residual = x

    x = layers.Conv2D(filters, 3, activation='relu', padding='same')(x)
    x = layers.Conv2D(filters, 3, activation='relu', padding='same')(x)

    ## MaxPooling2D를 사용할 경우
    if pooling:
        x = layers.MaxPooling2D(2, padding='same')(x)
        residual = layers.Conv2D(filters, 1, strides=2)(residual)

    ## MaxPooling2D를 사용하지 않을 경우
    elif filters != residual.shape[-1]:                 ## filters 갯수와 residual 모양이 다름
        residual = layers.Conv2D(filters, 1)(residual)

    x = layers.add([x, residual])

    return x

x = residual_block(x, filters=32, pooling=True)
x = residual_block(x, filters=64, pooling=True)
x = residual_block(x, filters=128, pooling=False)
## 마지막 블록 바로 전에 globalaverage pooling 사용으로 인해 pooling 필요하지 않음.

x = layers.GlobalAveragePooling2D()(x)

outputs = layers.Dense(1, activation='sigmoid')(x)

model = keras.Model(inputs=inputs, outputs=outputs)
model.summary()
keras.utils.plot_model(model, show_shapes=True)
```

>### 2) 배치 정규화
- 정규화(normalization): 머신 러닝 모델에 주입되는 샘플들을 균일하게 만드는 광범위한 방법
    - 사용 이유: 모델이 학습하고 새로운 데이터에 잘 일반화되도록 지원한다.
    - 방법
        - 1) (데이터 - 평균)을 빼서 데이터를 원점에 맞추고
        - 2) 표준 편차로 나누어 데이터의 분산을 1로 만든다.
        - Eq: (데이터 - 평균) / 표준편차
        - 데이터가 정규 분포(가우스)를 따른다고 가정하고 이 분포를 원점에 맞추고 분산이 1이 되도록 조정.        
- 데이터 정규화는 네트워크에서 일어나는 모든 변환 후에도 필요하다.(활성화 함수의 출력을 정규화)
- 아래와 같은 설계 순서를 따름
    - 배치 정규화가 입력 평균을 0으로 만들지만 relu 활성화 함수는 0을 기준으로 값을 통과시키거나 삭제(활성화 함수 이전에 정류화 수행 시, relu 함수 활용도 극대화)

```python
## 활성화 함수 지정을 BatchNormalization 층 다음에 설계
x = layers.Conv2D(32, 3, use_bias=False)(x)

x = layers.BatchNormalization()(x)

x = layers.Activation('relu')(x)
```

>### 3) 깊이별 분리 합성곱
- Conv2D를 대체하면서, 더작고 더 가벼운 모델 성능이 존재
    - 깊이별 분리합성곱
        - 공간 특성의 학습과 채널 방향 특성의 학습을 분리(채널 간에는 매우 독립적이기에 가능)     


```python
### 3가지 모두 적용하기 with ConvNet --> Xception 유사 모델.

## 데이터 증식단계 추가하기
data_augmentation = keras.Sequential(
    [
        layers.RandomFlip('horizontal'),
        layers.RandomRotation(0.1),
        layers.RandomZoom(0.2),
    ]
)


img_size=(180, 180, 3)
def get_model3():

    inputs = keras.Input(shape=img_size)

    x = data_augmentation(inputs)           ## 데이터 증식.
    x = layers.Rescaling(1./255)(x)         ## 입력 데이터 normalization

    ## RGB 채널은 높은 상관관계 --> 모델의 첫번째 층 일반적으로 Conv2D 설계
    ## 이 후에 SeparableConv2D 설계
    x = layers.Conv2D(filters=32, kernel_size=5, use_bias=False)(x)

    for size in [32, 64, 128, 256, 512]:

        residual = x

        x = layers.BatchNormalization()(x)
        x = layers.Activation('relu')(x)
        x = layers.SeparableConv2D(size, 3, padding='same', use_bias=False)(x)

        x = layers.BatchNormalization()(x)
        x = layers.Activation('relu')(x)
        x = layers.SeparableConv2D(size, 3, padding='same', use_bias=False)(x)

        x = layers.MaxPooling2D(3, strides=2, padding='same')(x)

        residual = layers.Conv2D(size, 1, strides=2, padding='same',
                                 use_bias=False)(residual)

        x = layers.add([x, residual])

    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.5)(x)

    outputs = layers.Dense(1, activation='sigmoid')(x)

    model = keras.Model(inputs=inputs, outputs=outputs)

    return model
```

```python
model = get_model3()
model.summary()
keras.utils.plot_model(model, show_shapes=True)
```

```python
import gdown
gdown.download(id='18uC7WTuEXKJDDxbj-Jq6EjzpFrgE7IAd', output='dogs-vs-cats.zip')

!unzip -qq dogs-vs-cats.zip
!unzip -qq train.zip

import os, shutil, pathlib

original_dir = pathlib.Path("train")
new_base_dir = pathlib.Path("cats_vs_dogs_small")


def make_subset(subset_name, start_index, end_index):
    for category in ("cat", "dog"):
        dir = new_base_dir / subset_name / category
        os.makedirs(dir)
        fnames = [f"{category}.{i}.jpg" for i in range(start_index, end_index)]
        for fname in fnames:
            shutil.copyfile(src=original_dir / fname,
                            dst=dir / fname)


make_subset("train", start_index=0, end_index=1000)
make_subset("validation", start_index=1000, end_index=1500)
make_subset("test", start_index=1500, end_index=2500)


from tensorflow.keras.utils import image_dataset_from_directory

train_dataset = image_dataset_from_directory(
    new_base_dir / "train",
    image_size=(180, 180),
    batch_size=32)
validation_dataset = image_dataset_from_directory(
    new_base_dir / "validation",
    image_size=(180, 180),
    batch_size=32)
test_dataset = image_dataset_from_directory(
    new_base_dir / "test",
    image_size=(180, 180),
    batch_size=32)
```

```python
epochs = 50

model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])

callbacks = [keras.callbacks.ModelCheckpoint(
    filepath='like_Xception.keras',
    save_best_only=True,
    monitor='val_loss')
    ]

history = model.fit(train_dataset,
                    epochs=epochs,
                    callbacks=callbacks,
                    batch_size=32,
                    validation_data=validation_dataset)
```

```python
import matplotlib.pyplot as plt

accuracy = history.history["accuracy"]
val_accuracy = history.history["val_accuracy"]
loss = history.history["loss"]
val_loss = history.history["val_loss"]
epochs = range(1, len(accuracy) + 1)

plt.plot(epochs, accuracy, "bo", label="Training accuracy")
plt.plot(epochs, val_accuracy, "b", label="Validation accuracy")
plt.title("Training and validation accuracy")
plt.legend()
plt.figure()

plt.plot(epochs, loss, "bo", label="Training loss")
plt.plot(epochs, val_loss, "b", label="Validation loss")
plt.title("Training and validation loss")
plt.legend()
plt.show()
```

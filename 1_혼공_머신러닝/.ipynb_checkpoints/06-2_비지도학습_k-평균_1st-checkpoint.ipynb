# 목표: k-평균 알고리즘의 작동 방식 이해, 과일 사진을 자동으로 모으기.

>### 06_1 절에서 과일들의 각 픽셀의 평균값 & Target을 통해 구분
>### 진짜 비지도 학습일 경우 어떻게 문제 해석 및 해결 => 어떻게 평균값을 유추할지

>### k-평균  군집 알고리즘을 이용하고 평균값 자동으로 추출.
>### 이 평균값이 클러스터의 중심에 위치, 때문에 클러스터 중심 또는 센트로이드

>1) 무작위로 k개의 클러스터 중심을 정한다.

>2) 각 샘플에서 가장 가까운 클러스터 중심을 찾아 해당 클러스터 샘플로 지정

>3) 클러스터에 속한 샘플의 평균값으로 클러스터 중심을 변경

>4) 클러스터 중심에 변화가 없을 때까지 2번으로 돌아가서 반복

### k-평균 모델 훈련하기 위해 3차원: (샘플 개수, 너비, 높이) => 2차원: (샘플 개수, 너비x높이)

```python
import numpy as np
fruits = np.load('fruits_300_data')

## 2차원으로 변경.
fruits_2d = fruits.reshape(-1, 100*100)

## k-평균, 설정할 매개변수: 클러스터 개수 지정.
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3, random_state=42)
km.fit(fruits_2d)

print(km.labels_)
```

>1) 위의 레이블 값 0,1,2 순서에는 어떠한 의미도 있지 않음.

>2) 어떠 레이블이 사과인지 확인할 수 있는 방법은 직접 이미지 출력이 최선.

```python
print(np.unique(km.labels_, return_counts=True))
```

### 각 클러스터가 어떤 이미지를 나타냈는지 그림으로 출력 함수 생성.

```python
import matplotlib.pyplot as plt

## 3차원 배열을 입력으로 받음 // figsize는 ratio 매개변수에 비례하여 커짐.
def func_draw_fruits(arr, ratio=1):
    
    # n은 샘플 개수
    n = len(arr)
    print(n)
    
    # 한 줄에 10개씩 이미지 그리기 // 샘플 개수를 10으로 나누어 전체 행 개수 계산
    # np.ceil 소수 첫번째 자리에서 올림.
    rows_cnt = int(np.ceil(n/10))
    print('행개수:', rows_cnt)
    
    # 행이 1개이면, 열의 개수는 샘플 개수. 그렇지 않으면 10개이다.
    cols_cnt = n if rows_cnt < 2 else 10
    print('열개수:', cols_cnt)
    
    fig, axs = plt.subplots(rows_cnt, cols_cnt, figsize=(cols_cnt*ratio, rows_cnt*ratio), squeeze=False)
    
    
    for i in range(rows_cnt):
        for j in range(cols_cnt):
            if i*10 + j < n:              # n개까지만 그린다.
                axs[i, j].imshow(arr[i*10 + j], cmap='gray_r')
            axs[i, j].axis('off')
    plt.show()    
```

```python
func_draw_fruits(fruits[km.labels_==0])
```

```python
func_draw_fruits(fruits[km.labels_==1])
```

```python
func_draw_fruits(fruits[km.labels_==2])
```

### KMeans 클래스가 최종적으로 찾은 클러스터 중심

```python
## 100 x 100 크기의 2차원 배열로 변경 필요.
func_draw_fruits(km.cluster_centers_.reshape(-1, 100, 100), ratio=3)
```

### 100번째 sample이 어떤 과일인지 알아보자.

```python
## 각 label에 따른 배열 크기 확인.
## KMeans 클래스는 가장 가까운 클러스터 중심을 예측하는 클래스
print(km.transform(fruits_2d[100:101]))
```

```python
print(km.predict(fruits_2d[100:101]))
```

-------------

### 1) 초기에 n_clusters = 3으로 지정, 타깃에 대한 정보 활용  실전에서는 클러스터 개수조차 알 수 없음.
### 2) 실전에는 n_clusters를 어떻게 지정해야 할 지 or 최적의 클러스터 개수?
### 3) k-평균 알고리즘 단점: 클러스터 개수를 사전에 지정해야 한다. => 엘보우 방법을 이용
### 4) (클러스터 중심 - 클러스터 내 샘플 사이)^2 들의 합 = 이너셔(inertia)
    a) 이너셔는 클러스터에 속한 샘플이 얼마나 가깝게 모여 있는지를 나타내는 값
    b) 일반적으로 클러스터 개수가 늘어나면 클러스터 개개의 크기는 줄어들기 때문에 이너셔도 줄어든다.
### 5) 엘보우 방법: 클러스터 개수를 늘려가면서 이너셔의 변화를 관찰하여 최적의 클러스터 개수를 찾는다.
    a) 최적의 클러스터 개수 k를 알아내는 한가지 방법은 클러스터가 얼마나 밀집되어 있는지 나타내는 방법 이너셔 사용
    b) 이너셔가 더 이상 크게 줄어들지 않는다면 클러스터 개수를 더 늘리는 것은 효과가 없다.
    c) 이너셔가 줄어드는 속도가 꺾이는 지점을 최적의 클러스터 개수로 결정.

```python
inertia = []
for k in range(2, 7):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(fruits_2d)
    inertia.append(km.inertia_)               ## KMeans 클래스, 자동으로 이너셔를 계산해서 itertia_ 속성으로 제공.

plt.plot(range(2, 7), inertia)
plt.xlabel('k')
plt.ylabel('inertia')
plt.show()
```

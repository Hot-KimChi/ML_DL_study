## Single Neuron Perceptron
- deep learning은 여러 개의 층으로 구성되어 있다.
- perceptron(혹은 neuron)은 층에 구성 요소이기에 이를 이해하는 것은 중요하다.



## Neuron의 안을 들여다 보면 아래와 같이 표현.

- x1, x2, x3는 inputs
- w1, w2, w3는 weight value
- 각각 x1 x w1 / x2 x w2 / x3 x w3 하여 더 해줌.
- bias는 최종 더한 값에 더하여 준다.

## Activation Function


```python
import tensorflow as tf
```

```python
## AND, OR, XOR operation 실행.
T = 1.0
F = 0.0
bias = 1.0
```

### 함수 구현

```python
def get_AND_data():
    X = [
    [F, F, bias],
    [F, T, bias],
    [T, F, bias],
    [T, T, bias]
    ]
    
    Y = [
        [F],
        [F],
        [F],
        [T]
    ]
    
    return X, Y

def get_OR_data():
    X = [
    [F, F, bias],
    [F, T, bias],
    [T, F, bias],
    [T, T, bias]
    ]
    
    Y = [
        [F],
        [T],
        [T],
        [T]
    ]
    
    return X, Y

def get_XOR_data():
    X = [
    [F, F, bias],
    [F, T, bias],
    [T, F, bias],
    [T, T, bias]
    ]
    
    Y = [
        [F],
        [T],
        [T],
        [F]
    ]
    
    return X, Y
```

```python
X, Y = get_AND_data()
```

### Weight 초기화
- 임의의 숫자로 가중치를 초기화

```python
W = tf.Variable(tf.random.normal([3, 1]))
```

### Activation function
- Perceptron은 step 함수를 activation function으로 구현.
- step(x) = { 1 if x > 0 ; 0 otherwise}

```python
def step(x):
    return tf.to_float(tf.greater(x, 0))
```

### Loss Function
- Mean Square Error로 loss function 구현

```python
f = tf.matmul(X, W)
output = step(f)
error = tf.subtract(Y, output)
mse = tf.reduce_mean(tf.square(error))
```

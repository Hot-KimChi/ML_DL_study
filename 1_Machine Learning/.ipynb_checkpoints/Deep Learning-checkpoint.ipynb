## Single Neuron Perceptron
- deep learning     .
- perceptron( neuron)       .



## Neuron      .

- x1, x2, x3 inputs
- w1, w2, w3 weight value
-  x1 x w1 / x2 x w2 / x3 x w3   .
- bias     .

## Activation Function


```python
import tensorflow as tf
```

```python
## AND, OR, XOR operation .
T = 1.0
F = 0.0
bias = 1.0
```

###  

```python
def get_AND_data():
    X = [
    [F, F, bias],
    [F, T, bias],
    [T, F, bias],
    [T, T, bias]
    ]
    
    Y = [
        [F],
        [F],
        [F],
        [T]
    ]
    
    return X, Y

def get_OR_data():
    X = [
    [F, F, bias],
    [F, T, bias],
    [T, F, bias],
    [T, T, bias]
    ]
    
    Y = [
        [F],
        [T],
        [T],
        [T]
    ]
    
    return X, Y

def get_XOR_data():
    X = [
    [F, F, bias],
    [F, T, bias],
    [T, F, bias],
    [T, T, bias]
    ]
    
    Y = [
        [F],
        [T],
        [T],
        [F]
    ]
    
    return X, Y
```

```python
X, Y = get_AND_data()
```

### Weight 
-    

```python
W = tf.Variable(tf.random.normal([3, 1]))
```

### Activation function
- Perceptron step  activation function .
- step(x) = { 1 if x > 0 ; 0 otherwise}

```python
def step(x):
    return tf.to_float(tf.greater(x, 0))
```

### Loss Function
- Mean Square Error loss function 

```python
f = tf.matmul(X, W)
output = step(f)
error = tf.subtract(Y, output)
mse = tf.reduce_mean(tf.square(error))
```

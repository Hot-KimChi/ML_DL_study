    - update한 값을 최종적으로 합하여 result에 update
- [A col] == 0 이 아닐 경우, 기존 [result ] 데이터 update

###  [A col]이 '0' 일 경우만 같은 번째에 있는 [option] 데이터를 list로 변환

```python
for i in range(len(data['A col'])):
    if data['A col'][i] == 0:
        list_option = data['option'].str.split(":")[i]
        print(list_option)
```

### 계산을 하기 위해, list를 floating으로 변환
- floating으로 변환 후, 절대값.

```python
for i in range(len(data['A col'])):
    
    if data['A col'][i] == 0:
        list_option = data['option'].str.split(":")[i]
        floatList = [float(x) for x in list_option]
        abs_value = np.abs(floatList)
        print(abs_value)
```

### 계산
- 1보다 클 경우, 뺀다
- 1보다 작을 경우, pass

```python
for i in range(len(data['A col'])):
    
    if data['A col'][i] == 0:
        list_option = data['option'].str.split(":")[i]
        floatList = [float(x) for x in list_option]
        abs_value = np.abs(floatList)
                
        calc =[]
        for value in abs_value:
            if 1 < value:
                calc.append(round(value-1, 4))
                
            else:
                calc.append(value)
        
        result = round(sum(calc), 2)
        print(result)
```

### [A col]이 '0' 일 아닐 경우, result값 유지

```python
list_result = []
for i in range(len(data['A col'])):
    
    if data['A col'][i] == 0:
        list_option = data['option'].str.split(":")[i]
        floatList = [float(x) for x in list_option]
        abs_value = np.abs(floatList)
                
        calc =[]
        for value in abs_value:
            if 1 < value:
                calc.append(round(value-1, 4))
                
            else:
                calc.append(value)
        
        result = round(sum(calc), 2)
        list_result.append(result)
    
    else:
        result = data['Result'][i]
        list_result.append(result)

print(list_result)
```
